```python
import pyarrow.parquet as pq
import pyarrow as pa
import os
```

```python

# Specify the Google Cloud Storage directory path
gcs_directory_path = 'mage-zoomcamp-zy/nyc_taxi_data'

# Set the path to your JSON key file
credentials_path = '/Users/kang/Documents/DE_zoomcamp/module_1_docker_terraform/gcp_terraform.json'

# Set the GOOGLE_APPLICATION_CREDENTIALS environment variable
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = credentials_path

# get the filesystem
gcs = pa.fs.GcsFileSystem()

# Create a ParquetDataset
dataset = pq.ParquetDataset(gcs_directory_path, filesystem=gcs)

# Read the dataset into a PyArrow Table
table = dataset.read(use_threads=True)

# Convert the Table to a Pandas DataFrame
df = table.to_pandas()

# Display the DataFrame
df
```
